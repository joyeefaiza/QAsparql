```python

#!pip uninstall pytorch_lightning
!pip install pytorch_lightning==1.7.2
#!pip install pytorch_lightning==1.1.5 torch==1.7.1
import pytorch_lightning as pl
from omegaconf import DictConfig, OmegaConf
import nemo
from nemo.collections.nlp.models.text2sparql import Text2SparqlModel
from nemo.core.config import hydra_runner

from nemo.utils import logging
from nemo.utils.exp_manager import exp_manager
```

```python
config_path = "/content/drive/MyDrive/Work/model/conf/text2sparql_config.yaml"
print(config_path)
config = OmegaConf.load(config_path)
print(OmegaConf.to_yaml(config))
```

```python
config.model.train_ds.filepath='/content/drive/MyDrive/Work/out/train2.tsv'
config.model.validation_ds.filepath='/content/drive/MyDrive/Work/out/test_easy.tsv'
config.model.test_ds.filepath='/content/drive/MyDrive/Work/out/test_hard.tsv'
config.model.batch_size=32 
config.model.nemo_path='/content/drive/MyDrive/Work/out/logs/bart.nemo' 
config.exp_manager.exp_dir='/content/drive/MyDrive/Work/out/logs'
cfg=config
```

```python
##############################Training the model over here as you can see we have fit functions and everything mentioned below

print("try")
t=""
tt=""
def main():
    global cfg
    global t
    global tt
    print("try")
    logging.info(f"Config:\n {cfg}")
    trainer = pl.Trainer(**cfg.trainer)
    exp_manager(trainer, cfg.get("exp_manager", None))
    nmt_model = Text2SparqlModel(cfg.model, trainer=trainer)
    t=trainer.fit(nmt_model)
    tt=nmt_model
    if cfg.model.nemo_path:
        nmt_model.save_to(cfg.model.nemo_path)
        print("run")


if __name__ == "__main__":
    main()
```
